1) Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a

Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.

```python
fig = px.histogram(tips, x = 'sex', y = 'total_bill',pattern_shape='smoker' , color='day')
```

```python
fig.show()
```

Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
the color parameter.

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

Q5. What is Distplot? Using Plotly express, plot a distplot.

The distplots in plotly python combines various statistical distribution such as histograms, violin plot, box plot etc in the same plot. To Create a distplots in plotly python, we can either use the px.histogram function from plotly express or use the create_distplot from the plotly figure_factory. 

```python
import plotly.express as px
fig = px.histogram(tips, x="sex", color="smoker",
                   marginal="violin")
fig.show()
```
